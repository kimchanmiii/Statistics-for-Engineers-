# 평균, 분산, 표준편차를 구하기 위한 함수를 작성한 후 math score, reading score, writing score에 대한 평균, 분산, 표준편차를 구하시오

```
import math

#extract column
math_score = df['math score']
reading_score = df['reading score']
writing_score = df['writing score']

#평균구하기 
def get_mean(lst):
  sum = 0
  for i in range(len(lst)):
    sum += lst[i]
  mean = sum / len(lst)
  return mean

print(get_mean(math_score), get_mean(reading_score), get_mean(writing_score))
print(np.mean(math_score), np.mean(reading_score), np.mean(writing_score))

#분산구하기 
def get_var(lst):
  sum = 0
  sum2 = 0
  for i in range(len(lst)):
    sum += lst[i]*lst[i]
    sum2 += lst[i]
  mean = sum2 / len(lst)
  var = (sum / len(lst)) - (mean*mean)
  return var

print(get_var(math_score), get_var(reading_score), get_var(writing_score))
print(np.var(math_score), np.var(reading_score), np.var(writing_score))

#표준편차구하기
def get_std(lst):
  sum = 0
  sum2 = 0
  for i in range(len(lst)):
    sum += lst[i]*lst[i]
    sum2 += lst[i]
  mean = sum2 / len(lst)
  var = (sum / len(lst)) - (mean*mean)
  std = math.sqrt(var)
  return std

print(get_std(math_score), get_std(reading_score), get_std(writing_score))
print(np.std(math_score), np.std(reading_score), np.std(writing_score))
```

중위수를 구하기 위한 함수를 작성한 후 math score, reading score, writing score에 대한 중위수를 구하시오

```
#중위수구하기
def get_median(lst):
    sortedLst = sorted(lst)
    lstLen = len(lst)
    index = (lstLen - 1) // 2

    if (lstLen % 2 == 0):
        return sortedLst[index]
    else:
        return (sortedLst[index] + sortedLst[index + 1])/2.0

print(get_median(math_score), get_median(reading_score), get_median(writing_score))
print(np.median(math_score), np.median(reading_score), np.median(writing_score))
```

female 및 male에 대한 math score의 평균, 분산, 표준편차를 구하시오

```
gender = df["gender"]

female_lst = []
male_lst = []

for i in range(1000):
  if gender[i] == "female":
    female_lst.append(math_score[i])
  elif gender[i] == "male":
    male_lst.append(math_score[i])

def get_mean(lst):
  sum = 0
  for i in range(len(lst)):
    sum += lst[i]
  mean = sum / len(lst)
  return mean

print(get_mean(female_lst), get_mean(male_lst))
print(np.mean(female_lst), np.mean(male_lst))
```
